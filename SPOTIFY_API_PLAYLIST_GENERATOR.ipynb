```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

def encode_genres_and_artist_value(df):
    # Extracting the genres and initializing genre columns
    genre_counts = {}
    for genres in df['Genres']:
        if pd.isnull(genres):
            continue
        for genre in str(genres).split(', '):
            genre_counts[genre] = genre_counts.get(genre, 0) + 1

    # Create a separate DataFrame to store genre encoding
    genre_df = pd.DataFrame(columns=genre_counts.keys(), index=df.index).fillna(0)

    # Filling the genre columns with appropriate values
    for index, row in df.iterrows():
        genres_in_row = str(row['Genres']).split(', ') if pd.notnull(row['Genres']) else []
        for genre in genres_in_row:
            genre_df.at[index, genre] = 1

    # Calculate the relative artist value
    artist_counts = df['Artists'].value_counts()
    df['relative artist value'] = df['Artists'].apply(lambda x: artist_counts[x])

    # Calculate the track relative value
    df['track relative value'] = df.index.to_series().apply(lambda index: (1 / (index + index + index+100)) + (df['relative artist value'].iloc[index] / ((index + index + index+100) / 10)))

    # Min-Max scale 'relative artist value' and 'track relative value'
    scaler = MinMaxScaler()
    df[['relative artist value', 'track relative value']] = scaler.fit_transform(df[['relative artist value', 'track relative value']])

    # Sort the columns in genre_df by their sum, from largest to smallest
    sorted_columns = genre_df.sum().sort_values(ascending=False).index.tolist()

    # Add the sorted genre columns to the original DataFrame
    for col in sorted_columns:
        df[col] = genre_df[col]

    # Reorder the columns to include "relative artist value" and "track relative value" in the desired locations
    columns_order = ['Track Name', 'track relative value', 'Album', 'Artists', 'relative artist value', 'Duration', 'Popularity', 'Added At', 'Genres'] + sorted_columns
    df=df.fillna(0)
    df = df[columns_order]

    return df




def get_top_tracks(encoded_dfs, n, genres=None):
    # Concatenate all the DataFrames in the list
    combined_df = pd.concat(encoded_dfs, ignore_index=True)

    # If genres are specified, filter the DataFrame to include only rows with the specified genres
    if genres:
        combined_df = combined_df[combined_df[genres].sum(axis=1) > 0]

    # Sort the combined DataFrame by "track relative value" in descending order
    sorted_df = combined_df.sort_values(by='track relative value', ascending=False)
    sorted_df=sorted_df.fillna(0)
    # Return the top n rows
    return sorted_df.head(n)



encoded_df = encode_genres_and_artist_value(df)
encoded_dfm = encode_genres_and_artist_value(dfm)
# print(encoded_df.head())
encoded_dfs=[encoded_df,encoded_dfm]

get_top_tracks(encoded_dfs, 10)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split

def split_and_reset_indexes(dfs, random_state=None):
    # Extract the DataFrames
    df1, df2 = dfs

    # Function to split a DataFrame into 4 equal parts
    def split_into_four(df):
        part1, temp = train_test_split(df, test_size=0.75, random_state=random_state)
        part2, temp = train_test_split(temp, test_size=0.6666, random_state=random_state)
        part3, part4 = train_test_split(temp, test_size=0.5, random_state=random_state)
        return part1, part2, part3, part4

    # Split both DataFrames into 4 parts
    df1_parts = split_into_four(df1)
    df2_parts = split_into_four(df2)

    # Reset the indexes of the DataFrames
    df_list = [*df1_parts, *df2_parts]
    for df in df_list:
        df.reset_index(drop=True, inplace=True)

    return df_list

# Usage
dfs = [df, dfm]
result_dfs = split_and_reset_indexes(dfs)

```
