##preprocessing

```python
from transformers import BertTokenizer, BertModel
import torch

from transformers import BertTokenizer, BertModel
import torch
import numpy as np
import spacy
import re
import pandas as pd
from textblob import TextBlob
from sklearn.base import BaseEstimator, TransformerMixin

class BertEmbedder(BaseEstimator, TransformerMixin):
    def __init__(self, model_name='bert-base-uncased'):
        self.tokenizer = BertTokenizer.from_pretrained(model_name)
        self.model = BertModel.from_pretrained(model_name)

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        embeddings = []
        for text in X:
            inputs = self.tokenizer(text, return_tensors='pt', truncation=True, padding=True)
            outputs = self.model(**inputs)
            embeddings.append(outputs.last_hidden_state[:, 0, :].cpu().detach().numpy())
        return np.concatenate(embeddings, axis=0)


class TextStats(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.stop_words = spacy.lang.en.stop_words.STOP_WORDS
        self.disaster_words = ["fire", "disaster", "flood", "earthquake", "hurricane", "tornado",
                               "tsunami", "avalanche", "typhoon", "drought", "blizzard", "hailstorm",
                               "cyclone", "epidemic", "pandemic", "inundation", "landslide",
                               "volcano", "wildfire", "accident", "crash", "explosion", "injury",
                               "collapse", "wreck", "eruption", "outbreak", "rescue", "evacuation",
                               "danger", "damage", "fatal", "casualty", "destruction", "tragedy",
                               "ruin", "hazard", "catastrophe", "calamity", "crisis", "violence",
                               "storm", "survivor", "emergency", "radiation", "pollution", "riot",
                               "war", "conflict", "terror", "bomb", "hostage", "assault", "shooting",
                               "california", "emergency", "buildings", "wildfire", "accident", "bombing",
                               "northern", "families", "suicide", "police", "killed", "crash", "fires",
                               "train", "storm", "hiroshima", "nuclear", "attack"]

        self.positive_emojis = set([":)", ":-)", ":D", ":d", "=)"])  # add more positive emojis here
        self.negative_emojis = set([":(", ":-(", ":c", "=(", ":<"])  # add more negative emojis here

    def fit(self, x, y=None):
        return self

    def transform(self, texts):
        data = []
        for text in texts:
            text_lowercase = text.lower()
            total_words = len(text.split())
            disaster_word_count = sum(text_lowercase.count(word) for word in self.disaster_words)
            data.append({
                'length': len(text),
                'num_words': total_words,
                'num_unique_words': len(set(w for w in text.split())),
                'words_vs_unique': len(set(w for w in text.split())) / len(text.split()),
                'num_stopwords': len([w for w in str(text_lowercase).split() if w in self.stop_words]),
                'num_hashtags': text.count('#'),
                'num_mentions': text.count('@'),
                'num_urls': len(re.findall(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', text)),
                'keyword_is_nan': int(pd.isnull(text)),
                'location_is_nan': int(pd.isnull(text)),
                'contains_time_value': int(bool(re.search(r'\b\d{1,2}:\d{2}\b|am|pm|utc', text, re.IGNORECASE))),
                'contains_distance_value': int(bool(re.search(r'\b\d+(\.\d+)?\s?(km|miles?|meters?)\b', text, re.IGNORECASE))),
                'disaster_word_count': disaster_word_count,
                'disaster_word_ratio': disaster_word_count / total_words if total_words > 0 else 0,
                'sentiment_polarity': TextBlob(text).sentiment.polarity,
                'sentiment_subjectivity': TextBlob(text).sentiment.subjectivity,
                'contains_number': int(any(char.isdigit() for char in text)),
                'num_exclamation_marks': text.count('!'),
                'num_question_marks': text.count('?'),
                'num_uppercase_chars': sum(1 for char in text if char.isupper()),
                'num_positive_emojis': sum(text.count(emoji) for emoji in self.positive_emojis),
                'num_negative_emojis': sum(text.count(emoji) for emoji in self.negative_emojis),
            })
        return data

def preprocess_data(df_train, df_test):
    df_train = df_train.fillna("")
    df_test = df_test.fillna("")

    bert_embedder = BertEmbedder()
    text_stats = TextStats()

    train_text_stats = pd.DataFrame(text_stats.transform(df_train['text']))
    test_text_stats = pd.DataFrame(text_stats.transform(df_test['text']))

    X_train_transformed = bert_embedder.transform(df_train['text'])
    X_test_transformed = bert_embedder.transform(df_test['text'])

    return X_train_transformed, X_test_transformed, train_text_stats, test_text_stats



# X_train_full, X_test_full, train_row_stats, test_row_stats = preprocess_data(dfTrain.drop('target', axis=1), dfTest)

```

```python
class ModelSelectionPipeline:
    def __init__(self, model, model_params, n_trials=100):
        self.model = model
        self.model_params = model_params
        self.n_trials = n_trials
        self.best_model = None
        self.best_predictions = None

    def objective(self, trial, X, y):
        params = self.model.get_params()  # Start with the model's default parameters
        for param, bounds in self.model_params.items():
            if isinstance(bounds[0], int):
                params[param] = trial.suggest_int(param, min(bounds), max(bounds))
            elif isinstance(bounds[0], float):
                params[param] = trial.suggest_float(param, min(bounds), max(bounds))
            else:
                params[param] = trial.suggest_categorical(param, bounds)
        self.model.set_params(**params)
        scores = cross_val_score(self.model, X, y, cv=5, scoring='accuracy')
        return scores.mean()

    def fit_and_predict(self, X_train, y_train, X_test):
        print(f"Evaluating model: {self.model.__class__.__name__}")
        study = optuna.create_study(direction='maximize')
        study.optimize(lambda trial: self.objective(trial, X_train, y_train), n_trials=self.n_trials)
        print(f"Best trial for model {self.model.__class__.__name__}: score {study.best_value}, params {study.best_params}")

        self.best_model = self.model.set_params(**study.best_params)
        self.best_model.fit(X_train, y_train)
        self.best_predictions = self.best_model.predict(X_test)


```

```python
#optunav2
class ModelSelectionPipeline:
    def __init__(self, model, model_params, n_trials=100, n_jobs=-1, pruner=optuna.pruners.MedianPruner()):
        self.model = model
        self.model_params = model_params
        self.n_trials = n_trials
        self.n_jobs = n_jobs
        self.pruner = pruner
        self.best_model = None
        self.best_predictions = None

    def objective(self, trial, X, y):
        params = self.model.get_params()  # Start with the model's default parameters
        for param, bounds in self.model_params.items():
            if isinstance(bounds[0], int):
                params[param] = trial.suggest_int(param, min(bounds), max(bounds))
            elif isinstance(bounds[0], float):
                params[param] = trial.suggest_float(param, min(bounds), max(bounds))
            else:
                params[param] = trial.suggest_categorical(param, bounds)
        self.model.set_params(**params)
        scores = cross_val_score(self.model, X, y, cv=10, scoring='accuracy', n_jobs=self.n_jobs)
        return scores.mean()

    def fit_and_predict(self, X_train, y_train, X_test):
        print(f"Evaluating model: {self.model.__class__.__name__}")
        study = optuna.create_study(direction='maximize', pruner=self.pruner)
        study.optimize(lambda trial: self.objective(trial, X_train, y_train), n_trials=self.n_trials, n_jobs=self.n_jobs)
        print(f"Best trial for model {self.model.__class__.__name__}: score {study.best_value}, params {study.best_params}")

        self.best_model = self.model.set_params(**study.best_params)
        self.best_model.fit(X_train, y_train)
        self.best_predictions = self.best_model.predict(X_test)



```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from scipy.sparse import hstack
from scipy.sparse import csr_matrix

def combine_features(matrix, dataframe):
    # Convert DataFrame to sparse matrix
    sparse_df = csr_matrix(dataframe.values)

    # Combine the two sparse matrices
    combined = hstack([matrix, sparse_df])

    return combined

def best_preds_one(models, param_grids, X_train_combined, X_test_combined, y, n_trials=50):
    # Split the combined training data into train and validation sets
    X_train, X_val, y_train, y_val = train_test_split(
        X_train_combined, y, test_size=0.1, random_state=42)

    test_predictions = []
    last_predictions = []


    for model, param_grid in zip(models, param_grids):
        pipeline = ModelSelectionPipeline(model, param_grid, n_trials=n_trials)
        pipeline.fit_and_predict(X_train, y_train, X_test_combined)


        pipeline_accuracy = accuracy_score(y_val, pipeline.best_model.predict(X_val))
        print(f"Meta model accuracy: {pipeline_accuracy}")

        pipeline.best_model.fit(X_val,y_val)
        pipeline_accuracy = accuracy_score(y_val, pipeline.best_model.predict(X_val))
        print(f"Meta model accuracy: {pipeline_accuracy} extra fitting")
        test_predictions.append(pipeline.best_model.predict(X_test_combined))
        print('------------------------------------------------------------------------')

        last_predictions=pipeline.best_model.predict(X_test_combined)


    return  test_predictions,last_predictions

# Call the function


X_train_combined = combine_features(X_train_full,  train_row_stats.drop(extra_columns, axis=1))
X_test_combined = combine_features(X_test_full, test_row_stats.drop(extra_columns, axis=1))

tt,last = best_preds_one(models, param_grids, X_train_combined, X_test_combined, dfTrain['target'])
```
